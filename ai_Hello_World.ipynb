# Hello world of Neural Networks
This is meant to guess the sequence of numbers y given x.
The actual relationship is y = 2*x-1. The neural network will try to guess this relationship given the information I feed it.

## Imports
This example uses tensorflow, numpy, and keras.
Keras defines a network as a set of sequential layers.

```
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

## Defining and Compiling the Neural Network
This is the most basic of neural networks - a fully connected network with 1 layer and 1 neuron.

```
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

The compiling stage defines a loss function and an optimizer algorithm to minimize that loss. The first run through the network gets a guess at the relationship between x and y. This is measured against the loss function, and the optimizer decides the best way to minimize that loss in the next pass.

```
model.compile(optimizer='sgd', loss='mean_squared_error')
```

## Provide the data
Declaring the data as a numpy array

```
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0
               , 1.0, 3.0, 5.0, 7.0], dtype=float)
```

## Training the Neural Network
The training process goes through a loop where the model makes a guess, measures how good/bad that guess is, goes through the optimization algorithm, and makes another guess. This process repeats for as many epochs as specified.

```
model.fit(xs, ys, epochs=500)
```

Making a prediction with the trained model:

```
print(model.predict([10.0]))
```

This doesn't predict 19 exactly because neural networks generate probabilities (as opposed to certainties). So the model gets close to the correct answer, but may not be exact.
